# Choosing a likelihood: Why negative binomial?
- count data
- positive only
- "poisson-like" but with overdispersion

```python
X = np.linspace(100, 900, 400).reshape(-1, 1)
f = np.exp(X * .005 - 2)
# y = st.nbinom.rvs(f, .1, loc=f)
y = np.random.negative_binomial(f, .5).ravel()

plt.scatter(X, y.astype(int))
# plt.scatter(X, f.astype(int))
plt.xlabel("Api Calls per second")
plt.ylabel("Stalls / s");
# plt.ylim(0, 1000)
```

$$\mu = exp(x*\beta + alpha)$$

$$y \sim NegBinomial( y | \mu, \phi)$$

$$Where$$

$$\mathop{{}\mathbb{E}}[y] = \mu$$ 

$$and$$
$$Var[y] = \mu + \frac{\mu^2}{\phi}$$

Phi basically describes the ratio between Var and Mean. For Poisson these are equal and there is no $\phi$

# (not) Choosing priors: How?
- Weakly informative priors and rescaling your data
- For ensuring conjugate priors please see:
    https://en.wikipedia.org/wiki/Conjugate_prior
- let's look at the code

```python
code = """
data {
    int N; //                ###########################################
    int<lower=0> y[N]; //    ## This is where all of the data comes   ##
    vector<lower=0>[N] x; // ## into the model. 
    int<lower=0> N_new; //   ## y, x, x_test                          ##
    vector[N_new] x_new; //  ###########################################
}
parameters {
    real beta;
    real alpha;
    real<lower=0> inv_phi;
}
transformed parameters {      // #######################################
    real<lower=0> phi;        // # more appropriate to put prior on    #
    phi <- 1 / sqrt(inv_phi); // # transformation of phi               #
}
model {
    beta ~ normal(0, 10); // <<<<< #######################################
    alpha ~ normal(0, 10); // <<<< ###### weakly informative priors ######
    inv_phi ~ normal(0, 10); // <<<<<< #######################################    

    // ######################################################
    // ###### Negative binomial likelihood and log link #####
    // ######################################################
    y ~ neg_binomial_2_log(x * beta + alpha, phi); // Generalized Linear Model with Negative Binomial Likelihood
}

// ######################################################
// ###### out of sample prediction for evaluation   #####
// ######################################################
generated quantities {
    vector[N_new] y_new;
    for (n in 1:N_new) {
        // random number generation of model fitted above (notice it's run on new data);
        y_new[n] = neg_binomial_2_log_rng(x_new[n] * beta + alpha, phi);
    }
}
"""
#################################
# COMPILE this is why it's fast #
#################################
model = pystan.StanModel(model_code=code)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y)
scaler = MinMaxScaler()
scaler.fit(x_train)


fit = model.sampling({
    'N': x_train.shape[0],
    'x': scaler.transform(x_train).ravel(),
    'y': y_train,
    'x_new': scaler.transform(x_test).ravel(),
    'N_new': x_test.shape[0]
})

fit
```

```python
fit.plot('beta');
fit.plot('alpha');
fit.plot('phi');
```

### Plot percentiles

```python
y_new = fit['y_new']
lp = fit['lp__']

for percentile in [99, 95, 80, 50, 20, 5, 1]:
    sort_mask = np.argsort(x_test.ravel())
    plt.plot(x_test.ravel()[sort_mask],
                np.percentile(y_new, percentile, axis=0)[sort_mask],
                label="{}th".format(percentile))
# plt.scatter(x_test[sort_mask], y_new[sort_mask])
# plt.scatter(x_train, y_train)
plt.scatter(x_test, y_test)
plt.legend()
plt.xlabel('Api Calls per second')
plt.ylabel('Stalls / s');
```
